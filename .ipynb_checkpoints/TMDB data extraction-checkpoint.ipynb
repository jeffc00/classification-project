# Movies

## Latest

```python
get = 'movie/latest'
response = requests.get(tmdb_api_url + '?'.join([get, api_key]))
latest_movie_id = response.json()['id']
```

## Details

```python
keys_to_keep = ['id', 'title', 'status', 'release_date', 'budget', 'revenue', 'genres',
                'original_language', 'runtime', 'popularity', 'vote_count', 'vote_average']
df_details = pd.DataFrame(columns=keys_to_keep)
movies = {}
for i in tqdm(range(latest_movie_id + 1)):
    get = f'movie/{i}'
    response = requests.get(tmdb_api_url + '?'.join([get, api_key]))
    if response.status_code == 200:
        result = response.json()
        result = {k: result[k] for k in keys_to_keep}
        result['genres'] = [d['name'] for d in result['genres']]
        row = pd.DataFrame.from_dict(result, orient='index').T
        df_details = df_details.append(row, ignore_index=True)
```
